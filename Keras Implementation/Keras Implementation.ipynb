```python
import numpy
import keras
# import tensorflow as tf
from keras.datasets import fashion_mnist
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from keras import optimizers
from keras.utils import np_utils
# with tf.device('/gpu:0'):
seed = 7
numpy.random.seed(seed)
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()
new_X_train=X_train[:6000]
new_Y_train=y_train[:6000]
new_X_test=X_test[:1000]
new_Y_test=y_test[:1000]
num_pixels = X_train.shape[1] * X_train.shape[2]
X_train = X_train.reshape(X_train.shape[0], num_pixels).astype('float32')
X_test = X_test.reshape(X_test.shape[0], num_pixels).astype('float32')
X_train = X_train / 255
X_test = X_test / 255
y_train = np_utils.to_categorical(y_train)
y_test = np_utils.to_categorical(y_test)
num_classes = y_test.shape[1]
from keras import optimizers
model = Sequential()
model.add(Dense(500, input_dim=784, kernel_initializer='normal', activation='relu'))
# model.add(Dense(num_classes, kernel_initializer='normal', activation='softmax'))

model.add(Dense(100, input_dim=500, kernel_initializer='normal', activation='relu'))
# model.add(Dense(num_classes, kernel_initializer='normal', activation='softmax'))

model.add(Dense(10, input_dim=100, kernel_initializer='normal', activation='sigmoid'))
# model.add(Dense(num_classes, kernel_initializer='normal', activation='softmax'))
# Compile model
gokul = keras.optimizers.Adam(lr=0.0001, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=False)
model.compile(loss='categorical_crossentropy', optimizer=gokul, metrics=['accuracy'])
bsize=int(input("Enter batch size: "))
num_iter = int(input("Enter number of epochs: "))
learning_rate = float(input("Enter the learning rate"))
# option = input('Enter which model you would like to test: 1)No momentum 2)Nestrov 3)RMSProp 4)Adam 5)Nadam')


#Initializing the gradient descent functions
rms=keras.optimizers.RMSprop(lr=learning_rate, rho=0.9, epsilon=None, decay=0.0)
gokul = keras.optimizers.Adam(lr=learning_rate, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=False)
nomomentum=keras.optimizers.SGD(lr=learning_rate, momentum=0.0, decay=0.0, nesterov=False)
nestrov=keras.optimizers.SGD(lr=learning_rate, momentum=0.0, decay=0.0, nesterov=True)
nadam = keras.optimizers.Nadam(lr=learning_rate, beta_1=0.9, beta_2=0.999, epsilon=None, schedule_decay=0.004)


gradients = [(nomomentum,"No Momentum"),(nestrov,"Nestrov"),(rms,"RMSProp"),(gokul,"Adam"),(nadam,"Nadam")]
for x in gradients:
    model = Sequential()
    model.add(Dense(500, input_dim=num_pixels, kernel_initializer='normal', activation='relu'))
    # model.add(Dense(num_classes, kernel_initializer='normal', activation='softmax'))

    model.add(Dense(100, input_dim=500, kernel_initializer='normal', activation='relu'))
    # model.add(Dense(num_classes, kernel_initializer='normal', activation='softmax'))

    model.add(Dense(10, input_dim=100, kernel_initializer='normal', activation='sigmoid'))
    # model.add(Dense(num_classes, kernel_initializer='normal', activation='softmax'))
    # Compile model
    model.compile(loss='categorical_crossentropy', optimizer=x[0], metrics=['accuracy'])
    vegeta=model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=num_iter, batch_size=bsize, verbose=2)
    # Final evaluation of the model
    print()
    print("Currently running "+ x[1] +" gradient function")
    print()
    scores = model.evaluate(X_test, y_test, verbose=0)
    print("Baseline Error: %.2f%%" % (100-scores[1]*100))
    #Plot of Errors
    import matplotlib.pyplot as plt
    plt.plot(vegeta.history['loss'])
    plt.plot(vegeta.history['val_loss'])
    plt.title(x[1]+' model loss')
    plt.ylabel('loss')
    plt.xlabel('epochs')
    plt.legend(['train', 'test'], loc='upper left')
    plt.show()
    #Plot of Accuracies
    plt.plot(vegeta.history['acc'])
    plt.plot(vegeta.history['val_acc'])
    plt.title(x[1] + ' model accuracy')
    plt.ylabel('accuracy')
    plt.xlabel('epoch')
    plt.legend(['train', 'test'], loc='upper left')
    plt.show()
    print()
    print()
```
